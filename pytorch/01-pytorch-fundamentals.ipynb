### 3.2 Matrix multiplication

Two ways of performing multiplication in neural network and deep learning:
1. Element-wise operation (broadcastin)
2. Matrix Multiplication (dot product)

Rules of matrix multiplication:
- `(m, n) @ (n, m) = (m, m)`

```python
# Element-wise multiplication
print(f"{tensor} * {tensor}")
print(f"Equals: {tensor*tensor}")
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor) # 1*1 + 2*2 + 3*3 = 14
```

```python
# Matrix multiplication using @
tensor @ tensor
```

```python
%%time

value = 0
for i in range(len(tensor)): # len(tensor) -> len(first dimension of tensor)
    value += tensor[i] * tensor[i]

print(value)
```

```python
%%time
torch.matmul(tensor, tensor)
```

## **4. One of the Most Common Errors in Deep Learning (Shape Errors)**

```python
# Shape of matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 8],
                         [9, 10],
                         [11, 12]])

torch.mm(tensor_A, tensor_B) # torch.mm is as the same as torch.matmul (it's an alias), but the input must be 2D
```

To fix tensor shape issues we can fix it by transposing it using `torch.transpose()` or `tensor.T`

A transpose switch the axes of a given tensor

```python
print(f"tensor_A:\n{tensor_A}\n")
print(f"tensor_B.T:\n{tensor_B.T}")
```

```python
torch.matmul(tensor_A, tensor_B.T)
```

### 4.4 Reshaping, stacking, squeezing, unsqueezing tensors
* Reshaping - reshapes an input tensor to a defined shape
* View - return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensor on top of each other (vstack) or side by side (hstack)
* Squeeze - remove all `1` dimension from a tensor
* Unsqueeze - add a `1` dimension from a tensor
* Permute - return a view of the input with dimensions permuted (swaped) in a certain way

```python
# Create a tensor
x = torch.arange(1., 11.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 10)
x_reshaped
```

```python
# Change the view
y = x.view(1, 10)
y, y.shape
```

```python
# Changing y changes x (because a view of tensor share the same memory as the original input)
y[0, 0] = 5.
y, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack((x, x, x), dim=0)
x_stacked
```

```python
x_stacked = torch.stack((x, x, x), dim=1)
x_stacked
```

```python
x_hstack = torch.hstack((x, x, x))
x_hstack
```

```python
x_vstack = torch.vstack((x, x, x))
x_vstack
```

```python
x_reshaped
```

```python
x_reshaped.shape
```

```python
# tensor.squeeze() remove all single dimension from a target tensor
print(f"Previous tensor\t: {x_reshaped}")
print(f"Previous shape\t: {x_reshaped.shape}")

# Remove extra dimension froms from x_reshaped
x_squeeze = x_reshaped.squeeze()
print(f"\nNew tensor\t: {x_squeeze}")
print(f"New shape\t: {x_squeeze.shape}")
```

```python
# tensor.unsqueeze() - adds a single dimension from a target tensor at a specific dim
print(f"Previous tensor\t: {x_reshaped}")
print(f"Previous shape\t: {x_reshaped.shape}")

# Add an extra dimension with unsqueeze
x_unsqueeze = x_reshaped.unsqueeze(dim=0) # axis = 0
print(f"New tensor\t: {x_unsqueeze}")
print(f"New shape\t: {x_unsqueeze.shape}")
```

```python
# tensor.permute() - rearranges the dimensions of the viewed tensor in a specific order (share the memory of the target tensor)
x_original = torch.rand(size=(224, 224, 3)) # [height, width, colour_channels]

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2 , 0 , 1) # shift axis 0 -> 1, axis 1 -> 2, 2 -> 0
print(f"Previous shape\t: {x_original.shape}")
print(f"New shape\t: {x_permuted.shape}") # [colour_channels, height, width]
```

```python
x_permuted[0,0,0] = 5
x_original[0,0,0], x_permuted[0,0,0] # they share the same memory
```

## **6. PyTorch Tensors and NumPy**

* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor to NumPy -> `torch.Tensor.numpy()`

```python
# NumPy array to tensor
import numpy as np

array = np.arange(1., 8.)
tensor = torch.from_numpy(array)
array, tensor
```

```python
array.dtype, torch.arange(1., 8.).dtype # default data of NumPy and PyTorch
```

```python
# Change the value of array, and see what it will do to `tensor`
array = array + 1
array, tensor
```

```python
# It will change the tensor too cause `+=` is an `in place` operation
array += 1
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Change the tensor, and what happen to numpy_tensor
tensor = tensor + 1
tensor, numpy_tensor
```

```python
tensor += 1
tensor, numpy_tensor
```

## **7. Reproducibility (Trying to Take Random Out of Random)**

To reduce the randomness in creating a random array or tensor comes the concept of `random seed`

```python
# Create 2 random tensor
tensor_A = torch.rand(3, 4)
tensor_B = torch.rand(3, 4)

print(tensor_A)
print(tensor_B)
print(tensor_A == tensor_B)
```

```python
# Let's make some random but reproducable tensor using `torch.manual_seed()`
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
tensor_D = torch.rand(3, 4)

print(tensor_C)
print(tensor_D)
print(tensor_C == tensor_D)
```
